## Moving Average

Moving average (rolling average or running average) is a calculation to analyze data points by creating a series of averages of different subsets of the full data set.


* blue curve : noizy sine  
* black curve : moving average

### Simple Moving Average
$ P_{0}$  : Current price  
$ P_{-t}$ : t days ago price

$$ SMA =  \frac{P_{0} + P_{-1} + \dots + P_{t-1}}{t} $$

### Weighted Moving Average
$ P_{0}$  : Current price  
$ P_{-t}$ : t days ago stock price


$$ SMA =  \frac{n \times P_{0} + (t-1) \times P_{-1} + \dots + 2 \times P_{t-2} + 1 \times P_{t-1}}{t + (t-1) + \dots + 2 + 1} $$


### Exponential Moving Average
$ P_{t}$ :  Current price  
$ P_{1}$  : t days ago stock price  


$${\displaystyle EMV_{t}={\begin{cases}P_{1},&t=1\\\alpha \cdot P_{t}+(1-\alpha )\cdot EMV_{t-1},&t>1\end{cases}}}$$
$${where\ \alpha = {\displaystyle \frac{smoothing}{(P+1)}}},\ {smoothing = 2}$$
