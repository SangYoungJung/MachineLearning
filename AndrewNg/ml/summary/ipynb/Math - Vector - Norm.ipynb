## Vector - Norm
The length of the vector, also called vector's magnitude  
  
  
$$\displaystyle \|{\boldsymbol {x}}\|_{p}:= (\sum _{i=1}^{n}\left|x_{i}\right|^{p})^{\frac{1}{p}}  $$    


### L1 Norm
The 1-norm is simply the sum of the absolute values of the columns.  
Taxicab norm or Manhattan norm  
$${\displaystyle \|{\boldsymbol {x}}\|_{1}:=\sum _{i=1}^{n}\left|x_{i}\right|}$$  
Example: 
* p=$\left[\begin{array}{ccc}
2 \\ 3 
\end{array}\right]$  L1 Norm = $|2|+|3| = 5$    
* p=$\left[\begin{array}{ccc}
3 \\ 1 \\ -3
\end{array}\right]
q=\left[\begin{array}{ccc}
5 \\ 2 \\ 6
\end{array}\right]
$ L1 Norm = $|3-5|+|1-2|+|-3-6| = 2 + 1 + 9 = 12 $  

**Note that :** L1 norm does not have a derivative. It is a nonsmooth curve.

### L2 Norm
Euclidean norm  
$${\displaystyle \|{\boldsymbol {x}}\|_{2}:= \sqrt{\sum _{i=1}^{n}x_{i}^{2}}} := (\sum _{i=1}^{n}x_{i}^{2})^{\frac{1}{2}} $$    
p=$\left[\begin{array}{ccc}
3 \\ 1 \\ -2
\end{array}\right]$ L2 Norm = $\sqrt{3^{2} + 1^{2} + (-2)^{2}} = \sqrt{14}$

### Maximum norm
If ${x}$ is some vector such that $x = (x_{1}, x_{2}, \dots, x_{n})$
$${\displaystyle \|{x}\|_{\infty}:=  max(x_{1}, x_{2}, \dots, x_{n})}$$
