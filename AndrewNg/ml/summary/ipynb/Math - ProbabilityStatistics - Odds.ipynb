## Probability - Odds


Conditional probability : The probability of A given B  
$$P(A \mid B) = \frac{P(A\ and\ B)}{P(B)}$$  
  
Binary Classificiation  
$ P(class1 \mid x) = y $  
$ P(class2 \mid x) = 1-y $  
$$ class1 \ \ if \ y > 0.5 $$
$$ class2 \ \ if \ y \le 0.5 $$


### Odds
$$ \frac{P}{1-P} > 1,\ if \ y > 0.5(class1)  $$
$$ [0,\infty] $$

### Loggit
The range [0, 1(or $\infty$]) to the full range of real numbers.  
$$ \ln{(\frac{P}{1-P})} > 0,\ if \ y > 0.5(class1)  $$
$$ [-\infty,\infty] $$

### Sigmoid(or Logistic function)
The inverse of the logit is the sigmoid. It is usefull at end of classification process.  
The full range of real numbers to the range [0, 1(or $\infty$]).
$$ logit(odds(p)) = \theta^{T}x = z $$
$\displaystyle \ln{(\frac{P}{1-P})} = z  $, $\displaystyle \frac{P}{1-P} = e^{z} $, $\displaystyle \frac{1}{P}-1 = \frac{1}{e^{z}} $, $\displaystyle \frac{1}{P} = \frac{1+e^{z}}{e^{z}} $  
$$ P = \frac{e^{z}}{1+e^{z}} = \frac{1}{1+e^{-z}} $$  


### Softmax
The softmax function(or normalized exponential function)  is a generalization of the logistic(or sigmoid) to multiple dimensions
$$Softmax(z_{i}) = \frac{e^{z_{i}}}{\sum_{j = 1}^{k} e^{zj}}$$  
  
Why $e$?
* Can use any number > 0 but consider last layer value as logit. it makes sure easy to calculate  
$$ e^{\ln{(\frac{P}{1-P})}} = \frac{P}{1-P}$$
