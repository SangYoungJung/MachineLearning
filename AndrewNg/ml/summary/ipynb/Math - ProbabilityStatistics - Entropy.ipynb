### Entropy
Entropy is a measure of the average amount of information.  
Random variable ${\displaystyle X}$ with probability mass function,  possible outcomes ${\displaystyle x_{1},...,x_{n}}$, probability ${\displaystyle \mathrm {P} (x_{1}),...,\mathrm {P} (x_{n})}$, entropy of ${\displaystyle X}$  
$\displaystyle H(X) = − \sum_{i=1}^{k}P(x_{i})log{P(x_{i})} = \sum_{i=1}^{k}P(x_{i})I(X) =  E[I(X)] = E[-log{P(x_{i})}] $  
$[0, 1] $  
by definition equal to the expected information content of measurement of ${\displaystyle X}$  


#### Example Coin flipping
Head Probability : $\frac{1}{2}$  
Tail Probability : $\frac{1}{2}$  
$\displaystyle H(x) = -(\frac{1}{2}log(\frac{1}{2})+\frac{1}{2}log(\frac{1}{2})) = 1 $

#### Drawing lots
* Suppose 1 red ball, 9 black balls  
$$\displaystyle H(x) = -(\frac{1}{10}log(\frac{1}{10})+\frac{9}{10}log(\frac{9}{10})) = 0.468 $$
* Suppose 2 red ball, 8 black balls  
$$\displaystyle H(x) = -(\frac{2}{10}log(\frac{2}{10})+\frac{8}{10}log(\frac{8}{10})) = 0.721 $$
* Suppose 5 red ball, 5 black balls  
$$\displaystyle H(x) = -(\frac{5}{10}log(\frac{5}{10})+\frac{5}{10}log(\frac{5}{10})) = 1.000 $$
  


### Cross - Entropy
Estimated probability distribution ${\displaystyle q}$, rather than the true distribution ${\displaystyle p}$.

$\displaystyle H(p,q) = − \sum_{i=1}^{k}p(x_{i})log{q(x_{i})}$  

#### Discreate(or Binary) distribution
Loss function of Classification  
$H(p)=-p\log _{2}q-(1-p)\log _{2}(1-q).$

### KL divergence(Kullback-Leibler)
${\displaystyle D_{\text{KL}}}$ (also called relative entropy), is a measure of how one probability distribution is different from a second.  
For discrete probability distributions ${\displaystyle P} and {\displaystyle Q}$ defined on the same probability space.  

${\displaystyle D_{\text{KL}}(P\parallel Q)=+\sum _{x\in {\mathcal {X}}}P(x)\log \left({\frac {P(x)}{Q(x)}}\right) =  -\sum _{x\in {\mathcal {X}}}P(x)\log \left({\frac {Q(x)}{P(x)}}\right)}$




* $KL(p|q) \ge 0$
* $KL(p|q) \ne KL(q|p)$
