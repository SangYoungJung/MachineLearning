## Convex and Concave function

### Convex Function
Convex if the line segment between any two points on the graph of the function does not lie below the graph between the two points.
  

| Line segment between any two points | Graph between the two points | 
| --- | --- |
| $ tf(x_{1}) + (1-t)f(x_{2}) $  | $ f(tx_{1} + (1-t)x_{2}) $ |  

$ {\displaystyle f:X\to \mathbb {R} } $  
$ {\displaystyle x_{1},x_{2}\in X} $ and $ t \in [0,1] $  
$ f(tx_{1} + (1-t)x_{2}) \le tf(x_{1}) + (1-t)f(x_{2}) $ 

#### Intuition with Vector
* v1=$\left[\begin{array}{ccc}
x1 \\ y1
\end{array}\right]$
* v2=$\left[\begin{array}{ccc}
x2 \\ y2
\end{array}\right]$

$ tv_{1} + (1-t)v_{2} = v_{2} + (v_{1} - v_{2})t $



if t = 0, $ v_{2} + (v_{1} - v_{2})t = v_{2} $  
if t = 1, $ v_{2} + (v_{1} - v_{2})t = v_{1} $  

### Concave Function
$ {\displaystyle f:X\to \mathbb {R} } $  
$ {\displaystyle x_{1},x_{2}\in X} $ and $ t \in [0,1] $  
$ f(tx_{1} + (1-t)x_{2}) \ge tf(x_{1}) + (1-t)f(x_{2}) $
