## Calculus - Limit

### Constant $e$
$\displaystyle \lim_{x \to 0} (1+x)^{\frac{1}{x}} = e$  

$\displaystyle \lim_{x \to \infty} \left(1+\frac{1}{x} \right)^{x} = e$  


### Limit of Exponential Functions and Logarithmic Functions

$\displaystyle \lim_{x \to 0} \frac{\ln{(1+x)}}{x} = 1$  

$\displaystyle \lim_{x \to 0} \frac{e^{x}-1}{x} = 1$  

$\displaystyle \lim_{x \to 0} \frac{\log_{a}({x}+1)}{x} = \frac{1}{\ln{(a)}}$  

$\displaystyle \lim_{x \to 0} \frac{a^{x}-1}{x} = \ln{(a)}$  

### To prove
$\displaystyle \lim_{x \to 0} \frac{e^{x}+1}{x} = 1$  

$\displaystyle \left[\begin{array}{}
t = e^{x}-1 \\
e^{x} = t+1 \\
x = \ln{(t+1)}, x \to 0, t \to 0 
\end{array}\right]$  

$\displaystyle \lim_{t \to 0} \frac{t}{\ln{(t+1)}} = \lim_{t \to 0} \frac{1}{\frac{1}{t}\ln{(t+1)}} = \frac{1}{\ln{(\lim_{t \to 0}(t+1)}^{\frac{1}{t}})} = \frac{1}{\ln(e)} = 1$
