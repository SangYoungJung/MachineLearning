## Calculus - Chain Rule
In calculus,  
the chain rule is a formula that expresses the derivative of the composition of two differentiable functions f and g in terms of the derivatives f and g. 

$ y=f(g(x)) $
$$ y'=f'(g(x))g'(x) $$  
Suppose $ y=f(z), z=g(x) $
$${\displaystyle {\frac {dy}{dx}}={\frac {dy}{dz}}\cdot {\frac {dz}{dx}}}$$

#### For Exmaple  
$ y = (2x^2 + 1)^2 $
$$ y' = 2(2x^2 + 1)^{2-1} * 4x^{2-1} $$
$$ y' = 16x^3 + 4x $$  

#### For Gradient descent
$$ h_{\theta}(x) = {\theta_{0}}x_{0}+{\theta_{1}}x_{1}+ ... + {\theta_{n}}x_{n}  $$  
$$ \theta(0,1 \sim  n)$$
$$x(0,1 \sim  n), x_0=1$$
$$y(1,2 \sim n)$$

$\displaystyle \frac{\partial}{\partial \theta_{j}} J({\theta}) = \frac{\partial}{\partial \theta_{j}} \frac{1}{2} (h_{\theta}(x)-y)^{2}$  
$\displaystyle = \frac{1}{2}*2(h_{\theta}(x)-y) * \frac{\partial}{\partial \theta_{j}}(h_{\theta}(x)-y)$  
$\displaystyle = (h_{\theta}(x)-y) * \frac{\partial}{\partial \theta_{j}}(\sum_{i = 0}^{n}{\theta_{i}x_{i}}-y)$  
$\displaystyle = (h_{\theta}(x)-y)x_{j}$


#### For **$\theta_{0}$** of Gradient descent  
$$ \frac{\partial}{\partial \theta_{0}} J({\theta_{0}}, {\theta_{1}}) = \frac{\partial} {\partial \theta_{0}} \frac{1}{2m} \sum_{i = 1}^{m} (h_{\theta}(x^{(i)})-y^{(i)})^{2}$$  
$$ h_{\theta}(x^{(i)}) = {\theta_{0}}+{\theta_{1}}x_{(i)} $$  

$\displaystyle = \frac{1}{2m} \sum_{i = 1}^{m} [(h_{\theta}(x^{(i)})-y^{(i)})^{2}]' * [(h_{\theta}(x^{(i)})-y^{(i)})]'$  
$\displaystyle = \frac{1}{2m} \sum_{i = 1}^{m} (h_{\theta}(x^{(i)})-y^{(i)}) * [({\theta_{0}}+{\theta_{i}}x^{(i)}-y^{(i)})]'$  
$\displaystyle = \frac{1}{m} \sum_{i = 1}^{m} (h_{\theta}(x^{(i)})-y^{(i)}) * (1)$  
$\displaystyle = \frac{1}{m} \sum_{i = 1}^{m} (h_{\theta}(x^{(i)})-y^{(i)})$
