### Calculus - Product  and Quotient Rule

### Product Rule
Using the power rule(or chain rule)  
$$((f + g)^{2})^{\prime} = 2(f + g)(f + g)^{\prime} = 2(ff^{\prime} + gg^{\prime} + fg^{\prime} + gf^{\prime})$$  
Expanding and then power rule  
$$((f + g)^{2})^{\prime} = (f^{2} + g^{2} + 2fg)^{\prime} = 2(ff^{\prime} +gg^{\prime} + {(fg)}^{\prime})$$  
Comparing above expressions,
$${(fg)}^{\prime} = fg^{\prime} + gf^{\prime}$$


### Quotient Rule
$\displaystyle \left[ \frac{1}{g(x)} \right]^{\prime} = \lim_{h \to 0}\frac{\frac{1}{f(g+h)}-\frac{1}{f(g)}}{h} $  
$\displaystyle = \lim_{h \to 0} \frac{g(x) - g(x+h)}{h\cdot g(x+h)\cdot g(x)}$  
$\displaystyle = \lim_{h \to 0} \left[ \frac{-1}{g(x+h)\cdot g(x)} \cdot \frac{g(x+h)-g(x)}{h} \right]$  
$\displaystyle = \lim_{h \to 0} \frac{-1}{g(x+h)\cdot g(x)} \cdot \lim_{h \to 0} \frac{g(x+h)-g(x)}{h}$  
$\displaystyle = \frac{-1}{{g(x)}^{2}} \cdot g(x)^{\prime}$
$\displaystyle = \frac{-g(x)^{\prime}}{{g(x)}^{2}}$
